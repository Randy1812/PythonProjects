```python
import pandas as pd
import plotly.express as px
```

```python
df_apps = pd.read_csv("./apps.csv")
```

```python
df_apps.sample(10)
```

```python
df_apps.drop(['Last_Updated','Android_Ver'],axis=1,inplace=True)
df_apps.sample(10)
```

```python
df_apps_clean = df_apps.dropna()
df_apps_clean
```

```python
dupl_rows = df_apps_clean[df_apps_clean.duplicated()]
dupl_rows
```

```python
df_apps_clean.drop_duplicates(subset=["App","Type","Price"])
```

```python
df_apps_clean.sort_values('Rating',ascending=False).head()
```

```python
df_apps_clean.sort_values('Size_MBs',ascending=False).head()
```

```python
df_apps_clean.sort_values('Reviews',ascending=False).head()
```

```python
rating = df_apps_clean.Content_Rating.value_counts()
rating
```

```python
fig = px.pie(labels=rating.index, values=rating.values, title="Content Rating", names=rating.index, hole=0.6)
fig.update_traces(textposition="outside", textinfo="percent+label")
```

```python
df_apps_clean.Installs.describe()
```

```python
df_apps_clean.info()
```

```python
df_apps_clean[["App","Installs"]].groupby("Installs").count()
```

```python
df_apps_clean.Installs = df_apps_clean.Installs.astype(str).str.replace(',','')
df_apps_clean.Installs = pd.to_numeric(df_apps_clean.Installs)
df_apps_clean[["App","Installs"]].groupby("Installs").count()
```

```python
df_apps_clean.info()
```

```python
df_apps_clean.Price.describe()
```

```python
df_apps_clean.Price = df_apps_clean.Price.astype(str).str.replace("$","")
df_apps_clean.Price = pd.to_numeric(df_apps_clean.Price)
df_apps_clean[["App","Price"]].sort_values("Price",ascending=False).head(20)
```

```python
df_apps_clean = df_apps_clean[df_apps_clean["Price"]<250]
df_apps_clean.sort_values("Price", ascending = False).head(5)
```

```python
df_apps_clean["Revenue Estimate"] = df_apps_clean.Installs.mul(df_apps_clean.Price)
df_apps_clean.sort_values("Revenue Estimate", ascending=False)[:10]
```

```python
df_apps_clean.Category.nunique()
```

```python
top10_category = df_apps_clean.Category.value_counts()[:10]
top10_category 
```

```python
bar = px.bar(x=top10_category.index, y=top10_category.values)
bar.update_layout(xaxis_title="Category", yaxis_title="Count")
```

```python
category_installs = df_apps_clean.groupby('Category').agg({"Installs":pd.Series.sum})
category_installs.sort_values("Installs",ascending=True, inplace=True)
category_installs
```

```python
h_bar = px.bar(x=category_installs.Installs, y=category_installs.index, orientation='h', title="Category Popularity Distribution")
h_bar.update_layout(xaxis_title="Number Of Installs", yaxis_title="Category")
```

```python
df_apps_clean
```

```python
cat_number = df_apps_clean.groupby("Category").agg({"App":pd.Series.count})
```

```python
cat_merged_df = pd.merge(cat_number, category_installs, on="Category", how="inner")
print(f"The dimensions of the Dataframe are : {cat_merged_df.shape}")
cat_merged_df.sort_values("Installs",ascending=False)
```

```python
scatter = px.scatter(cat_merged_df, x="App", y="Installs", title="Category Concentration",size="App", hover_name=cat_merged_df.index,color="Installs")
scatter.update_layout(xaxis_title="Number Of Apps (Lower = More Concentrated)", yaxis_title="Installs", yaxis=dict(type="log"))
```

```python
len(df_apps_clean.Genres.unique())
```

```python
df_apps_clean.Genres.value_counts().sort_values(ascending=True)[:5]
```

```python
stack = df_apps_clean.Genres.str.split(';',expand=True).stack()
print(f"We now have a single column with shape: {stack.shape}")
num_genres = stack.value_counts()
print(f"Number of genres : {len(num_genres)}")
```

```python
genre_bar = px.bar(x=num_genres.index[:15], y=num_genres.values[:15], color=num_genres.values[:15], color_continuous_scale="agsunset")
genre_bar.update_layout(coloraxis_showscale=False, xaxis_title="Genres", yaxis_title="Number Of Apps")
```

```python
df_apps_clean.Type.value_counts()
```

```python
df_free_vs_paid = df_apps_clean.groupby(["Type","Category"], as_index=False).agg({"App": pd.Series.count})
df_free_vs_paid
```

```python
app_type_bar = px.bar(df_free_vs_paid, x="Category", y="App", title="Free Vs Paid Apps by Category", color="Type", barmode="group")
app_type_bar.update_layout(xaxis_title="Category", yaxis_title="Number Of Apps", xaxis={'categoryorder':'total descending'}, yaxis=dict(type='log'))
```

```python
box = px.box(df_apps_clean, y="Installs", x="Type", color="Type", notched=True, points='all', title="How Many Downloads Paid Apps Are Giving Up?")
box.update_layout(yaxis=dict(type='log'))
```

```python
df_paid_apps = df_apps_clean[df_apps_clean['Type'] == 'Paid']
box = px.box(df_paid_apps, 
             x='Category', 
             y='Revenue Estimate',
             title='How Much Can Paid Apps Earn?')
 
box.update_layout(xaxis_title='Category',
                  yaxis_title='Paid App Ballpark Revenue',
                  xaxis={'categoryorder':'min ascending'},
                  yaxis=dict(type='log'))
```

```python
box = px.box(df_paid_apps,
             x='Category',
             y="Price",
             title='Price per Category')
 
box.update_layout(xaxis_title='Category',
                  yaxis_title='Paid App Price',
                  xaxis={'categoryorder':'max descending'},
                  yaxis=dict(type='log'))
 
```
